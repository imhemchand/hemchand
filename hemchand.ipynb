```python
def func():
    n=int(input())
    sum=0
    if(n%2==0):
        for i in range(1,n):
            if(n%i==0):
                sum+=i
        if(sum==n):
            print("perfect")
        else:
            print("not perfect")
        
    else:
        fact=1
        for i in range(1,n+1):
            fact*=i
        print("the number is odd,fact is",fact)
func()
```
